```python
#temp=dataMovie['genres'].str.split('|',n=4,expand = True)
#dataMovie['genres'][0]
#len(dataMovie.loc[:,'genres'])
#bool=dataMovie['genres'].map(lambda x:len(x)).value_counts()
#dataMovie[dataMovie['genres']=='(no genres listed)'].count()
lista=pd.DataFrame(["Action","Adventure","Animation","Comedy","Crime","Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi","Thriller","War","Western","(no genres listed)"])


#dataMovie[dataMovie['genres'].str.contains('Adventure')].value_counts()

lista['count']=0
def count_genres():
    for i in range(len(lista)):
        val=dataMovie['genres'].str.contains(lista.loc[i,0]).value_counts()
        #lista['count'][i]=val[True]
        lista.loc[i,'count']=int(val[val.index==True].reset_index(drop=True)[0])
    lista.sort_values('count', inplace=True, ascending=True)
        
count_genres()
lista.plot(0,'count',kind='barh')
```

```python
temp=dataMovie['title'].str.extract(r'(.{7}$)')
temp=temp[0].str.extract(r'([0-9]+)')  
dataMovie['year_premiere']=temp
dataMovie.head()
```

```python
dataMovie[dataMovie['year_premiere'].isnull()]
```

```python
dataMovie['year_premiere'].fillna(0, inplace=True)
dataMovie['year_premiere'].isnull().sum()


```

```python
dataMovie['year_premiere']=dataMovie['year_premiere'].astype(int)

año_estreno=dataMovie['year_premiere'].value_counts()
año_estreno.sort_index(inplace=True)
año_estreno.plot(kind='barh', figsize=[5,20],ylim =[1914,2016])
```

```python
#dataRatings['año']=
#date=temp['timestamp'].apply(datetime.fromtimestamp)
#date.year
dataRatings.loc[:,'date']=pd.to_datetime(dataRatings.loc[:,'timestamp'],unit='s')
dataRatings.loc[:,'year_rating']=dataRatings['date'].map(lambda x:x.year)
dataRatings.head()
```

```python
dataRatings.loc[:,'month']=dataRatings['date'].map(lambda x:x.month)
dataRatings.head(20)
```

```python
#dataMerge=pd.merge(dataRatings, dataMovie, on='movieId')
#dataMerge.head()
#datetime.fromtimestamp(temp.loc[1,'timestamp'])
#import matplotlib
promYear=dataRatings.groupby(['year_rating'])['rating'].mean()
promMonth=dataRatings[dataRatings['year_rating']>2010].groupby(['month'])['rating'].mean()
```

```python
countRating=dataRatings.groupby(['movieId'], as_index=False)['rating'].count()
countRating#movieId  rating
#arraylimit=countRating[countRating['rating']>500].movieId
#arraylimit
dataMerge=pd.merge(dataRatings, countRating, on='movieId')
dataRatings=dataMerge[dataMerge['rating_y']>500]
dataRatings
```

```python
#año=dataRatings['timestamp'].map(lambda x:datetime.fromtimestamp(x).year)
#año
#import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(4,8))
#fig.set_ylim([0,3.6])
ax.bar(promYear.index,promYear.values)
```

```python
#temp.loc[:,'timestamp']
#temp=dataRatings.head()
#dataRatings.append(año,ignore_index=True)
#promMonth=dataRatings[dataRatings['year']>2010].groupby(['month'])['rating'].count()
promMonth=dataRatings[dataRatings['year_rating']>2010].groupby(['month'])['rating_y'].count()
#fig, ax1 = plt.subplots(figsize=(4,6))
#ax1.bar(promMonth.index,promMonth.values)
promMonth.plot(kind='bar')
```

```python
#temp['timestamp'].map(lambda x:datetime.fromtimestamp(x).year)
ratingMean=dataRatings.groupby('movieId',as_index=False)['rating_x'].mean()
ratingMean.sort_values('rating_x', inplace=True)
ranking=pd.merge(ratingMean, dataMovie, on='movieId')
ranking
```
