```python
generos = df_pelicula.genres.str.split(pat="|")
array_genero=[]
for rowgenero in generos:
    for gene in rowgenero:
        array_genero.append(gene)
genero_unico_df = pd.Series(array_genero).unique()
cantidad_genero = genero_unico_df.shape[0]
genero_unico_df
```

```python
index = [0]
genero_pelicula_df = pd.DataFrame(index=index, columns=genero_unico_df)
genero_pelicula_df.fillna(0, inplace=True)
for gu in genero_unico_df:
    for rowgenero in generos:
        for gene in rowgenero:
            if gu == gene:
                genero_pelicula_df[gu][0] += 1
##valorgenero = df_pelicula.genres.value_counts()
##valorgenero
genero_pelicula_df
```

```python
genero_pelicula_df_vertical = pd.DataFrame(index=genero_unico_df, columns=["cantidad"])
genero_pelicula_df_vertical.fillna(0, inplace=True)
for gu in genero_unico_df:
    for rowgenero in generos:
        for gene in rowgenero:
            if gu == gene:
                genero_pelicula_df_vertical["cantidad"][gu] += 1
genero_pelicula_df_verticalcallable
```

```python
x = np.arange(20)
print(x)
plt.bar(x, height= genero_pelicula_df_vertical.cantidad)
plt.xticks(x, genero_unico_df)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = genero_unico_df
students = genero_pelicula_df_vertical.cantidad.values
ax.bar(langs,students)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = genero_unico_df
students = genero_pelicula_df_vertical.cantidad.values
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```
